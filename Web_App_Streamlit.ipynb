```python
%%writefile app.py
import streamlit as st
import pandas as pd

# -------------------------------
# 1Ô∏è‚É£ Page Configuration
# -------------------------------
st.set_page_config(
    page_title="Quiz App",
    page_icon="üß†",
    layout="centered"
)

st.title("üß† Python Web Quiz App")
st.write("Test your Python & Data Knowledge")

st.divider()

# -------------------------------
# 2Ô∏è‚É£ USER INPUT (Professional Touch)
# -------------------------------
name = st.text_input("üë§ Enter your name")

st.divider()

# -------------------------------
# 3Ô∏è‚É£ QUIZ QUESTIONS
# -------------------------------
questions = [
    {
        "question": "Which language is best for data analysis?",
        "options": ["HTML", "Python", "CSS", "PHP"],
        "answer": "Python"
    },
    {
        "question": "Which library is used for data visualization?",
        "options": ["NumPy", "Pandas", "Matplotlib", "Flask"],
        "answer": "Matplotlib"
    },
    {
        "question": "Which is the correct file extension for python files?",
        "options": [".pyth", ".pt", ".py", ".pyt"],
        "answer": ".py"
    },
    {
        "question": "Which keyword is used to define a function in python?",
        "options": ["function", "define", "def", "fun"],
        "answer": "def"
    },
     {
        "question": "which symbol is used for comments in python?",
        "options": ["//", "/**/", "#", "<!-- -->"],
        "answer": "#"
    },
    {
        "question": "Which of the following is a valid variable name?",
        "options": ["2name", "my-name", "my_name", "my name"],
        "answer": "my_name"
    },
     {
        "question": "How do you take input from the user?",
        "options": ["input()", "read()", "scan()", "get()"],
        "answer": "get()"
    },
    {
        "question": "What is the output of this code? print(type(10))",
        "options": ["<class 'float'>", "<class 'str'>", "<class 'int'>", "<class 'bool'>"],
        "answer": "<class 'int'>"
    },
    {
        "question": "How do you create a list in python?",
        "options": ["{1, 2, 3}", "(1, 2, 3)", "[1, 2, 3]", "<1, 2, 3>"],
        "answer": "[1, 2, 3]"
    },
    {
        "question": "What keyboard is used for a conditional statement?",
        "options": ["if", "when", "check", "case"],
        "answer": "if"
    },
    {
        "question": "What will this code output? print(10 > 5)",
        "options": ["True", "False", "10 > 5", "Error"],
        "answer": "True"
    },
    {
        "question": "What function gives the length of a list?",
        "options": ["len()", "count()", "size()", "length()"],
        "answer": "len()"
    },
    {
        "question": "Which symbol is used for equality comparison?",
        "options": ["=", "==", "!=", "<="],
        "answer": "<="
    },
    {
        "question": "Which does range(3) produce?",
        "options": ["1, 2, 3", "0, 1, 2", "0, 1, 2, 3", "1, 2"],
        "answer": "0, 1, 2"
    },
    {
        "question": "Which keyword is used to create a class?",
        "options": ["object", "define", "class", "struct"],
        "answer": "class"
    },
     {
        "question": "Which function convert a value to an interger?",
        "options": ["str()", "int()", "float()", "bool()"],
        "answer": "int()"
    },
    {
        "question": "Which framework builds fast web apps?",
        "options": ["Django", "Streamlit", "TensorFlow", "Keras"],
        "answer": "Streamlit"
    }
]

score = 0

# -------------------------------
# 4Ô∏è‚É£ QUIZ LOGIC + PROGRESS BAR
# -------------------------------
progress = st.progress(0)

for i, q in enumerate(questions):
    st.subheader(f"Q{i+1}. {q['question']}")
    answer = st.radio(
        "Choose an option:",
        q["options"],
        key=i
    )

    if answer == q["answer"]:
        score += 1

    progress.progress((i + 1) / len(questions))

# -------------------------------
# 5Ô∏è‚É£ SUBMIT & RESULTS
# -------------------------------
if st.button("üöÄ Submit Quiz"):

    if name.strip() == "":
        st.warning("Please enter your name before submitting.")
    else:
        percentage = (score / len(questions)) * 100

        st.success(f"Well done {name}! üéâ")
        st.write(f"Score: **{score}/{len(questions)}**")
        st.write(f"Percentage: **{percentage:.2f}%**")

        # Pass / Fail status
        if percentage >= 60:
            st.balloons()
            st.success("Status: PASS ‚úÖ")
        else:
            st.error("Status: FAIL ‚ùå")

        # -------------------------------
        # 6Ô∏è‚É£ RESULT VISUALIZATION
        # -------------------------------
        chart_df = pd.DataFrame({
            "Result": ["Correct", "Wrong"],
            "Count": [score, len(questions) - score]
        })

        st.bar_chart(chart_df.set_index("Result"))

        # -------------------------------
        # 7Ô∏è‚É£ SAVE RESULT (Optional)
        # -------------------------------
        result = pd.DataFrame([{
            "Name": name,
            "Score": score,
            "Total": len(questions),
            "Percentage": percentage
        }])

        result.to_csv("quiz_results.csv", mode="a", header=False, index=False)
```
