```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# App Title
st.title("ðŸŽ“ Student Marks Analyzer")

st.write("Enter marks for each subject to analyze performance")

# Input fields
name = st.text_input("Student Name")

subjects = ["Maths", "Science", "English", "Computer", "Social"]
marks = []

for subject in subjects:
    mark = st.number_input(f"{subject} Marks", min_value=0, max_value=100)
    marks.append(mark)

# Button
if st.button("Analyze Marks"):
    # Create DataFrame
    df = pd.DataFrame({
        "Subject": subjects,
        "Marks": marks
    })

    total = sum(marks)
    percentage = total / len(subjects)

    # Grade logic
    if percentage >= 90:
        grade = "A+"
    elif percentage >= 75:
        grade = "A"
    elif percentage >= 60:
        grade = "B"
    elif percentage >= 40:
        grade = "C"
    else:
        grade = "Fail"

    # Display results
    st.subheader(f"ðŸ“Œ Student: {name}")
    st.write(df)

    st.success(f"Total Marks: {total}")
    st.success(f"Percentage: {percentage:.2f}%")
    st.success(f"Grade: {grade}")

    # Bar Graph
    st.subheader("ðŸ“Š Marks Distribution")
    fig, ax = plt.subplots()
    ax.bar(df["Subject"], df["Marks"])
    ax.set_ylabel("Marks")
    ax.set_title("Marks per Subject")
    st.pyplot(fig)
```
